**The loss function**
The loss function is a measurement of error which defines the precision lost on comparing the preicted output to the actual output


loss = [(actual output) - (predicted output)]^2


The weight that cross x as 0 is the one with the smallest loss

The aim is to reach a point where the slope is zero

Positive slope = increase in weight
Negative slope = decrease in weight
Zero slope = approximate weight
